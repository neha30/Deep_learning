## Regression with Keras

Data: Boston House price dataset

nn_model.png


For our use case we are considering Boston house price dataset. We have 13 input columns and one target column.

Lets consider NN consist of one input layer, two hidden layer and one output layer. We have 13 input columns so our input layer will have 13 neurons and lets consider 4 neurons in each hidden layers. you can change the number of neurons in the hidden layer, for simplicity we are taking only 4 neurons. 



```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataframe = pd.read_csv("data/housing.csv", delim_whitespace=True, header=None)
```

```python
dataframe.head()
```

```python
# divide the data into input and target columns
dataset = dataframe.values
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
X.shape
```

Because our model contain linear stack of layers, we will use keras 'Sequential'.  There are two models in the Keras library. One of them is the Sequential model and the other one is the model class used with the functional API. So to create your model, you simply call the Sequential constructor. 

```python
model = Sequential()
n_cols = X.shape[1]
no_neurons = 4
model.add(Dense(no_neurons, activation='relu', input_shape=(n_cols,)))
model.add(Dense(no_neurons, activation='relu'))
model.add(Dense(1))

#now for training define optimizer and loss function
#adam is more efficient then gradient descent for deep learning. One advantage of adam is, we don't need to
# specify the learning rate.
model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(X, Y)
```

```python
model.predict()
```
