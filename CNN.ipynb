## Convolutional Neural Network

CNN are similar to neural network.

CNN takes input as a image.

Solves problem involving Image recognition, object detection and other computer vision applicaiton.









```python
import keras
from keras.models import Sequential
from keras.datasets import mnist
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D

%matplotlib inline
```

```python
#download mnist data and split into train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
#plot the first image in the dataset
plt.imshow(X_train[0])
```

```python
X_train.shape
```

```python
#check image shape
X_train[0].shape
```

```python
#reshape data to fit model
#(no. on images, shape 1 of iamge, shape2 of image, 1 if image is grey scale)
X_train = X_train.reshape(60000,28,28,1)
X_test = X_test.reshape(10000,28,28,1)
```

```python
# one hot encode the target variable
#one-hot encode target column
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
y_train[0]
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, kernel_size=(2,2), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(20, activation= 'relu'))
model.add(Dense(10, activation='softmax'))

# sftmax convert the o/p into probability
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5)
```
